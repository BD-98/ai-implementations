```python
import numpy as np 

from sklearn.linear_model import LogisticRegression as LR
from sklearn import datasets
from sklearn.model_selection import train_test_split
np.set_printoptions(suppress=True)
```

**Steps**

1. Initialize: l_rate, number of iters, weight, bias
2. Fit(x,y): Define the hypothesis and update the weights 
3. Predict(X): Return the hypothesis with the updatted weights

**Formulae**

dw = $\frac{1}{m}X^T[h_\theta(x) - y]$

w = w - l_rate*dw


```python
# Logistic Regression Implementation 

class LogisticRegression:
    def __init__(self, lr=0.001, iters=1000):
        self.lr = lr 
        self.iters = iters 
        self.weight = None 
    
    def fit(self,X,y):
        # init params 
        n_samples, n_features = X.shape 
        self.weight = np.zeros(n_features)

        for i in range(self.iters):
            hypothesis = X.dot(self.weight)
            y_pred = self._sigmoid(hypothesis)
            dw = (1/n_samples) * (X.T.dot(y_pred - y))
            self.weight -= self.lr*dw 
    
    def predict(self, X):
        hypothesis = X.dot(self.weight)
        y_pred = self._sigmoid(hypothesis)
        return np.where(y_pred>0.5,1,0)
        
    def _sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    

    
```

```python
# Date Preproccessing 
bc = datasets.load_breast_cancer()
X, y = bc.data, bc.target 
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.5, random_state=1234)

# Checking Our Model Accuracy
accuracy = lambda y_true, y_pred: (y_true == y_pred).mean()    
model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)


# Comparing to Sci-kit Learn 
test_model = LR()
test_model.fit(x_train, y_train)
y_pred_test = test_model.predict(x_test)
scikit_model_accuracy = accuracy(y_test, y_pred_test) 
my_model_accuracy = accuracy(y_test, y_pred)
scikit_model_accuracy, my_model_accuracy
```
